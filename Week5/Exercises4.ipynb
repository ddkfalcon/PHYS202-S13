```python
from math import factorial
from pylab import *
x = arange(-4., 4.01, 1e-3)

def Hn(n,x):
    H0 = Hlast2 = 1
    H1 = Hlast = 2 * x
    if n == 0:
        return H0
    elif n == 1:
        return H1
    else:
        H = 0
        for k in range(1, n):
            H = Hlast * 2 *x - 2 * k * Hlast2
            Hlast2 = Hlast
            Hlast = H
        return H

n = 0
H = Hn(n,x)
psi_0 = 1 / sqrt(2 ** n * factorial(n) * sqrt(pi)) * exp(-x ** 2 / 2) * H

n = 1
H = Hn(n,x)
psi_1 = 1 / sqrt(2 ** n * factorial(n) * sqrt(pi)) * exp(-x ** 2 / 2) * H

n = 2
H = Hn(n,x)
psi_2 = 1 / sqrt(2 ** n * factorial(n) * sqrt(pi)) * exp(-x ** 2 / 2) * H

n = 3
H = Hn(n,x)
psi_3 = 1 / sqrt(2 ** n * factorial(n) * sqrt(pi)) * exp(-x ** 2 / 2) * H

plot_0 = plot(x, psi_0, lw = 3, label = 'n = 0')
plot_1 = plot(x, psi_1, lw = 3, label = 'n = 1')
plot_2 = plot(x, psi_2, lw = 3, label = 'n = 2')
plot_3 = plot(x, psi_3, lw = 3, label = 'n = 3')
xlabel(r'$x$', fontsize = 20)
ylabel(r'$\psi_n(x)$', fontsize = 20)
xlim(-4., 4.)
legend()
show()
```
