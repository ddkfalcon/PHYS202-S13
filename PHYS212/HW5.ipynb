```python
T = arange(0., 1e5, 0.1)
Eu = 1
kb = 8.62e-5 
Eavg = lambda T: Eu / (exp(Eu / (kb * T)) + 1)

plot(T, Eavg(T), lw = 3, label = r"$\bar{E} = \frac{E_u}{e^{\frac{E_u}{k_B T}} + 1}, E_u = 1$")
title(r"$\bar{E}$ vs $T$", fontsize = 18)
xlabel("Temperature (K)")
ylabel("Average Energy")
legend(loc = 0, prop = {'size':25})
show()
```
