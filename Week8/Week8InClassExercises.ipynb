Example 1: Analytic Inverse CDF Transform
-

```python
x = arange(0., pi + 0.001, 0.01)
f = lambda x: sin(x) / 2.
F = lambda x: (cos(0) - cos(x)) / 2.

plot(x, f(x), label = 'True distribution', lw = 3)
plot(x, F(x), label = "Cumulative function", lw = 3)

xlim(0, 3.5)
ylim(0, 1.1)

legend(loc = 0)
show()
```

```python
#number of samples
N = 10000

#N random numbers between 0 and 1
z = numpy.random.random_sample(N);

#inverse CDF
Finv = lambda z: arccos(cos(0) - 2 * z)
    
#sampled x values
xsamples = Finv(z);

#how do they compare to the original?
hist(xsamples, 50, normed = True, label = "Sampled distribution")
plot(x, f(x), label = 'True distrubution', lw = 3, color = 'red')

xlim(0, pi)
ylim(0, 0.7)

legend(loc = 0)
show()
```

Exercise 2. Exponential function - numerical

(a)

```python
dist = f(x)
CDF = cumsum(dist) / sum(dist)
F_inverse = x[searchsorted(CDF, n)]
```

(b)

(c)

```python
def distribute1D(x, prob, N):
    """takes any numerical distrubution prob on the interval defined by the array x,
    and returns an array of N sampled vallues that are statistically
    the same as the input data."""
    dist = prob(x)
    CDF = cumsum(dist) / sum(dist)
    n = numpy.random.random(N)
    F_inverse = x[searchsorted(CDF, n)]
    return F_inverse
```

Exercise 3. Quantum double-slit experiment
-

(a)

```python
d = 15e-6
a = 10e-6
L = 1
lmbda = 632.8e-9

sinc = lambda x: sin(pi * x) / (pi * x)
I_diff = lambda x: sinc((a * x) / (lmbda * L)) ** 2
I_int = lambda x: cos((2 * pi * d * x) / (lmbda * L)) ** 2
I_tot = lambda x: I_diff(x) * I_int(x)
```

(b)

```python
x = arange(-0.2, 0.201, 0.001)
dist = distribute1D(x, I_tot, N)
n,bins,patches = hist(dist, 400, normed = True, label = 'Sample distribution')
plot(x, I_tot(x) * max(n), 'r')
xlim(-0.2, 0.2)
legend()
show()
```
