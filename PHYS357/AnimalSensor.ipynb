The graphs are not raw data.  They are calibrated to centimeters.  The pseudocode for conversion is as follows:  
    
    sum = 0
    averageRange = 60  
    for i from 0 to averageRange, i++  
        analogVolt = analogRead(analogPin) / 2  
        sum += analogVolt  
    inches = sum / averageRange  
    centimeters = inches * 2.54  



Five Days Of Data
-

```python
"""Importing the data and sorting it"""

t, dist = loadtxt('DISTANCE.TXT', unpack = True, delimiter = ',', usecols = [0, 1])

tDay1 = t[0:17]
distDay1 = dist[0:17]

tDay2 = t[18:24]
distDay2 = dist[18:24]

tDay3 = t[25:46]
distDay3 = dist[25:46]

tDay4 = t[47:89]
distDay4 = dist[47:89]

tDay5 = t[90:len(t)]
distDay5 = dist[90:len(dist)]
```

Plotting The Five Days
-

Day One:

```python
plot(tDay1, distDay1, 'b.', label = "Day 1", lw = 3)
title("Animal Distance vs. Time", size = 18)
xlabel("$Time (ms)$", size = 14)
ylabel("$Distance (cm)$", size = 14)
legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)
show()
```
