The graphs are not raw data.  They are calibrated to centimeters.  The pseudocode for conversion is as follows:  
    
    sum = 0
    averageRange = 60  
    for i from 0 to averageRange, i++  
        analogVolt = analogRead(analogPin) / 2  
        sum += analogVolt  
    inches = sum / averageRange  
    centimeters = inches * 2.54  
    
A distance is only logged if something moves in front of the sonar.  
Pseudocode:  
  
    save last distance reading  
    compare to current distance reading  
    if last distance is not equal to current distance  
    record the distance

Horse Distance Sensor
-

After we logged our five days of data, we had a bit of time left.  So we decided to log more data, in a slightly different context.  Ms. Totten owns two horses which are free to roam around in a roughly 600 square feet fenced area.  We pointed the sensor toward the horses, to log the positions and the movements of the horses.

```python
"""Importing the data and sorting it"""

tHorse, distHorse = loadtxt('DISTANCE2.TXT', unpack = True, delimiter = ',', usecols = [0, 1])
```

```python
plot(tHorse, distHorse, 'g*', label = 'Horse Distance')
title("Horse Distance vs. Time", size = 18)
xlabel("$Time (ms)$", size = 14)
ylabel("$Distance (cm)$", size = 14)
legend(loc = 'best')
show()
```

Analysis
-

We can see that the position is more dense closer to the sensor.  Judging from this data, we can see that the horses enjoy being by the fence (where the sensor is).  Perhaps because this is where Ms. Totten enters to feed and interact with her horses.  The points become quite dense around 35,000 ms.  Maybe at this point, Ms. Totten entered the fenced area and played with her horses or dropped off some food.
