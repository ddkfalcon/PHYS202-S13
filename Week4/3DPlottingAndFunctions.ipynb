Electric Dipole Equation
=

\begin{equation}
V(x,y) = \frac{kq}{\sqrt{x^2 + (y - d)^2}} - \frac{kq}{\sqrt{x^2 + (y + d)^2}}
\end{equation}

```python
#make the x-y grid to cover the plane -2 < x < 2 and -2 < y < 2
dx = 0.06
dy = 0.06
x = arange(-2., 2., dx)
y = arange(-2., 2., dy)
print x.ndim
print len(x)

```

```python
X, Y = meshgrid(x,y)
print X.ndim
print Y.shape
print X
print Y
```

```python
#make the function, V(x,y)
k = 1; q = 1; d = 1
V = (k*q/(X**2 + (Y - d)**2)**(1/2.)) - (k*q/(X**2 + (Y + d)**2)**(1/2.))
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, V, rstride = 1, cstride = 1, linewidth = 0, cmap = cm.winter_r)
#help(Axes3D.plot_surface)
cb = fig.colorbar(p, shrink = 0.5)
```

```python
#Represent 3D data by contours
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])
contour(X, Y, V, levels, linewidths = 2, cmap = cm.winter_r)
show()
```

Ex 2. Electric Potential
-

```python
dx = 8./101.
dy = 8./101.
x = arange(-4., 4., dx)
y = arange(-4., 4., dy)
X,Y = meshgrid(x,y)
k = 8.9875517873681764*(10**9) #N m^2 C^-2
q = 10**-9 #Coulombs
d = 2 #meters
V = (k*q/(X**2 + (Y - d)**2)**(1/2.)) + (k*q/(X**2 + (Y + d)**2)**(1/2.))
U = -((k*q/(Y**2 + (X - d)**2)**(1/2.)) + (k*q/(Y**2 + (X + d)**2)**(1/2.)))
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, V+U, rstride = 1, cstride = 1, linewidth = 0, cmap = cm.prism_r)
cb = fig.colorbar(p, shrink = 0.5)
```

```python
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])
contour(X, Y, V + U, levels, linewidths = 2, cmap = cm.winter_r)
show()
```

```python
def pointPotential(x, y, q, posx, posy):
    #Returns electric potential due to given charge
    #k = 8.9875517873681764*(10**9) #N m^2 C^-2
    k = 1
    X,Y = meshgrid(x, y)
    Vxy = (k*q/((X - posx)**2 + (Y - posy)**2)**(1/2.)) #- (k*q/(X**2 + (Y + posy)**2)**(1/2.))
    return Vxy
dx = 0.06
dy = 0.06
x = arange(-2., 2., dx)
y = arange(-2., 2., dy)
V11 = pointPotential(x, y, 1, 1., 1.)
print V11
```

```python
def dipolePotential(x, y, q, d):
    #returns the dipole potential 
    k = 8.9875517873681764*(10**9) #N m^2 C^-2
    X,Y = meshgrid(x,y)
    Vxy = (k*q/(X**2 + (Y - d)**2)**(1/2.)) + (k*q/(X**2 + (Y + d)**2)**(1/2.))
    Uxy = -((k*q/(Y**2 + (X - d)**2)**(1/2.)) + (k*q/(Y**2 + (X + d)**2)**(1/2.)))
    return Vxy + Uxy
dx = 8./101.
dy = 8./101.
x = arange(-4., 4., dx)
y = arange(-4., 4., dy)
V22 = dipolePotential(x,y, 1e-9, 2.)
print V22
```
