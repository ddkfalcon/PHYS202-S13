```python
y = arange(-5.00, 5.01, 0.01)
E = cosh(y)
p = sinh(y)
```

```python
import matplotlib.pyplot as plt
ax1 = plt.axes(frameon = False)
ax1.set_frame_on(False) #constructing axis
ax1.get_xaxis().tick_bottom() #disable drawing ticks on top
#ax1.axes.get_yaxis().set_visible(False) #turn off y axis
plot(y, E, 'bx')
plot(y, abs(p), 'ro')
ax1.xaxis.labelpad = 30
```
