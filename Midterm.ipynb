```python
ft = lambda t: (t + 1) / (t - 1)
gt = lambda t: exp(-t)
t = arange(-5., 5.01, 0.001)
plot(t, ft(t), color = 'blue', lw = 3, label = "(1+t)/(1-t)")
plot(t, gt(t), color = 'red', linestyle = ':', lw = 3, label = "exp(-t)")
ylim(-10, 10)
xlim(-4.5, 4.5)
xlabel(r'time (seconds)', fontsize = 20)
ylabel(r'f(t), g(t)', fontsize = 20)
legend(loc = 3)
show()
```

```python
#writing code here then copy pasting
def pointField(x,y,q,Xq,Yq):
    k = 1 #N m^2 C^-2
    Ex =  k * q * (x - Xq) / ((x - Xq) ** 2 + (y - Yq) ** 2) ** (1 / 2)
    Ey =  k * q * (y - Yq) / ((x - Xq) ** 2 + (y - Yq) ** 2) ** (1 / 2)
    Exy = [sum(Ex), sum(Ey)]
    return Exy
#Imported into nano
```

```python
def is_anagram(s1,s2):
    if len(s1) == len(s2):
        test = []
        for letter in s1:
            for letter2 in s2:
                if letter == letter2:
                    test.append(True)
        if len(test) == len(s1):
            return True
        else:
            return False
    else:
        return False
```

```python
s1 = "mud"
s2 = "dum"
result = is_anagram(s1, s2)
print result
```
